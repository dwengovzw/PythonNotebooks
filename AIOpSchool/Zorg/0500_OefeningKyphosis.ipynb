```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import tree
```

```python
## Oefening 'Kyphosis'
```

```python
kyphosis = pd.read_csv("data/kyphosis.csv") 
```

```python
kyphosis
```

```python
# categorische variabelen numeriek maken
kyphosis["Kyphosis"]= kyphosis["Kyphosis"].replace("absent", 0)
kyphosis["Kyphosis"]= kyphosis["Kyphosis"].replace("present", 1)
```

```python
kyphosis
```

```python
kyphosis = np.array(kyphosis)
```

```python
kyphosis
```

```python
# beschouwde parameters zijn 'leeftijd', 'aantal' en 'start'
# waarden van deze parameters staan in resp. kolom 1, 2 en 3 van matrix
# elke rij komt overeen met één kind
# eerste kolom geeft of persoon kyphosis heeft ('1') of niet ('0').
```

```python
# parameters en klasse onderscheiden
parameters = kyphosis[:, 1:]        # laatste 3 kolommen van matrix zijn beschouwde parameters
klasse = kyphosis[:, 0]             # laatste kolom zijn klasse waartoe persoon behoort
```

```python
print(parameters)
```

```python
print(klasse)
```

```python
# beslissingsboom genereren op basis van data
beslissingsboom = tree.DecisionTreeClassifier(criterion="gini")   # boom wordt aangemaakt via gini-index
beslissingsboom.fit(parameters, klasse)                           # boom genereren die overeenkomt met data
```

```python
plt.figure(figsize=(20,15))  
tree.plot_tree(beslissingsboom, 
               class_names=["geen kyphosis", "kyphosis"], 
               feature_names=["Leeftijd", "Aantal", "Start"], 
               filled=True, rounded=True)
plt.show()
```
