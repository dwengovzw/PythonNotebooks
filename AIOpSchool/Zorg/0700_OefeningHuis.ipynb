### Voorbeeldoplossing

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
# beschouwde parameters zijn 'ingrijpende operatie', 'familie thuis' en 'oud'
# alle parameters zijn categorisch, met waarden 'ja' en 'nee'
# we zetten ze om naar numerieke 
# 'nee' is '0' en 'ja' is '1'
# waarden van deze parameters staan in resp. kolom 0, 1 en 2 van matrix data
# elke rij komt overeen met één patiënt
# laatste kolom geeft of patiënt na één dag naar huis mag ('1') of niet ('0').
data = np.array(
      [[0, 3, 1, 1],
       [1, 3, 0, 0],
       [0, 4, 0, 1],
       [0, 3, 0, 0],
       [1, 4, 0, 1]])
```

```python
# parameters en klasse onderscheiden
parameters = data[:, :3]        # eerste 3 kolommen van matrix zijn beschouwde parameters
klasse = data[:, 3]             # laatste kolom zijn klasse waartoe persoon behoort
```

```python
print(parameters)
```

```python
print(klasse)
```

```python
# beslissingsboom genereren op basis van data
beslissingsboom = tree.DecisionTreeClassifier(criterion="gini")   # boom wordt aangemaakt via gini-index
beslissingsboom.fit(parameters, klasse)    # boom genereren die overeenkomt met data
```

```python
plt.figure(figsize=(15,10))  
tree.plot_tree(beslissingsboom, 
               class_names=["komt niet in aanmerking", "komt in aanmerking"], 
               feature_names=["Gemeubeld", "Aantal slaapkamers", "Nieuwe keueken"], 
               filled=True, rounded=True)
plt.show()
```
